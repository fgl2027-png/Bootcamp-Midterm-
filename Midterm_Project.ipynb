```python
df.set_index('Series Description', inplace=True)
```

```python
df.head()
```

```python
df = df.apply(pd.to_numeric, errors='coerce') # makes strings into floats
df = df.replace(',', '', regex=True)
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
df = df.rename(columns={
    "Total consumer credit owned and securitized, not seasonally adjusted level": "Total",
    "Revolving consumer credit owned and securitized, seasonally adjusted level": "Revolving",
    "Nonrevolving consumer credit owned and securitized, seasonally adjusted level": "Nonrevolving",
    "Percent change of total consumer credit, seasonally adjusted at an annual rate":"% Change in Total",
    "Percent change of total revolving consumer credit, seasonally adjusted at an annual rate":"% Change in Revolving",
    "Percent change of total nonrevolving consumer credit, seasonally adjusted at an annual rate":"% Change in Nonrevolving",
    "Revolving consumer credit owned and securitized, not seasonally adjusted level":"Revolving credit owned",
    "Nonrevolving consumer credit owned and securitized, not seasonally adjusted level":"Nonrevolving credit owned",
    "Total consumer credit owned and securitized by finance companies, not seasonally adjusted level":"Total owned by finance companies",
    "Total consumer credit securitized by depository institutions, not seasonally adjusted level":"Total owned by depository institutions"
})
```

```python
df["Revolving Share"] = df["Revolving"] / df["Total"]
df["Nonrevolving Share"] = df["Nonrevolving"] / df["Total"]
df["Total Growth"] = df["Total"].pct_change()
```

```python
df["Total"].plot(title="Total Consumer Debt Over Time")
```

```python
df[["Revolving", "Nonrevolving"]].plot(title="Debt by Type")
```

```python
df[["Revolving Share", "Nonrevolving Share"]].plot(title="Debt Composition")
```

```python
df["Total Growth"].plot(title="Growth Rate of Total Debt")
```
